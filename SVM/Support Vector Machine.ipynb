```python
import pandas as pd
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
```

```python
iris = load_iris()
iris
```

```python
iris.feature_names
```

```python
dir(iris)
```

```python
df = pd.DataFrame(iris.data,columns=iris.feature_names)
df
```

```python
df["target"] = iris.target
df
```

```python
df["flower name"] = df.target.apply(lambda x : iris.target_names[x])
df
```

```python
df0 = df[df.target==0]
df1 = df[df.target==1]
df2 = df[df.target==2]
```

### Plot the graph to see the classification 

```python
plt.xlabel('sepal length (cm)')
plt.ylabel('sepal width (cm)')
plt.scatter(df0['sepal length (cm)'],df0['sepal width (cm)'],color='red',marker='*')
plt.scatter(df1['sepal length (cm)'],df1['sepal width (cm)'],color='blue',marker='*')
plt.scatter(df2['sepal length (cm)'],df2['sepal width (cm)'],color='green',marker='*')
```

```python
plt.xlabel('petal length (cm)')
plt.ylabel('petal width (cm)')
plt.scatter(df0['petal length (cm)'],df0['petal width (cm)'],color='red',marker='*')
plt.scatter(df1['petal length (cm)'],df1['petal width (cm)'],color='blue',marker='*')
plt.scatter(df2['petal length (cm)'],df2['petal width (cm)'],color='green',marker='*')
```

### Split the data

```python
from sklearn.model_selection import train_test_split
```

```python
X = df.drop(['target','flower name'],axis='columns')
X
```

```python
y = df['target']
y
```

```python
X_train, X_test , y_train , y_test = train_test_split(X,y,train_size=0.8,random_state=10)
```

```python
X_test
```

```python
from sklearn.svm import SVC
```

```python
model = SVC()
model
```

```python
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```
