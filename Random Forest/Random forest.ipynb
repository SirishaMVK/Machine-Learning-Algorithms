### Random Forest
we divide the dataset in random sets and then we build a decision tree for each dataset and majority output of those is

```python
import pandas as pd
from sklearn.datasets import load_digits
import matplotlib.pyplot as plt
```

```python
digits = load_digits()
digits
```

```python
plt.gray()
for i in range(4):
    plt.matshow(digits.images[i])
```

```python
df =pd.DataFrame(digits.data, columns=digits.feature_names)
df
```

```python
df["target"] = digits.target
df
```

```python
X = df.drop("target",axis="columns")
X
```

```python
y = df.target
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train , X_test , y_train , y_test = train_test_split(X, y,test_size=0.2)
```

### Ensemble is to call mutliple algorithms, here Random forest is using many decision trees 

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier()
```

```python
model.fit(X_train, y_train)
```

```python
#estimator is how many decision trees are used 
```

```python
model.score(X_test, y_test)
```

### Confusion matrix 

```python
y_predicted = model.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
cm = confusion_matrix(y_test, y_predicted)
cm
```

```python
import seaborn as sns
plt.xlabel("y_predicted")
plt.ylabel("y_test")
sns.heatmap(cm, annot = True)
```
