```python
import pandas as pd
```

```python
df = pd.read_csv("titanic.csv")
df.head(5)
```

```python
df.columns
```

```python
df.dtypes
```

```python
df.isnull().sum()
```

```python
df[["Age"]]=df[["Age"]].fillna(df.Age.median())
df
```

```python
df.isnull().sum()
```

```python
input1 = df.drop(["PassengerId","Survived","Name","SibSp","Parch","Ticket","Cabin","Embarked"],axis="columns")
input1
```

```python
result=df["Survived"]
result
```

```python
#inputs.Sex = inputs.Sex.map({'male': 1, 'female': 2})
```

### Label Encoder for sex as it a categorical variable

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le_sex = LabelEncoder()
```

```python
input1["sex_n"] = le_sex.fit_transform(input1["Sex"])
input1
```

### drop the sex column 

```python
input_n = input1.drop(["Sex"],axis="columns")
input_n
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(input_n,result,train_size=0.8, random_state=10)
```

```python
X_train
```

```python
y_train
```

### Import decesion tree model

```python
from sklearn import tree
```

```python
model = tree.DecisionTreeClassifier()
model
```

```python
model.fit(X_train, y_train)
```

```python
model.predict(X_test)
```

```python
model.predict([[3,28.5,7.2292,1]])
```

```python
model.score(X_test, y_test)
```
