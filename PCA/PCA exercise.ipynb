### Pls convert the dataset with category dtypes in data that is compatibe for machine learning

```python
df3.dtypes
```

### Sex , ChestPainType, RestingECG, ExerciseAngina,ST_Slope are object datasets 

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le_Sex = LabelEncoder()
le_ChestPainType = LabelEncoder()
le_RestingECG = LabelEncoder()
le_ExerciseAngina = LabelEncoder()
le_ST_Slope = LabelEncoder()
```

```python
df3['Sex'] = le_Sex.fit_transform(df3['Sex'])
df3['ChestPainType'] = le_ChestPainType.fit_transform(df3['ChestPainType'])
df3['RestingECG'] = le_RestingECG.fit_transform(df3['RestingECG'])
df3['ExerciseAngina'] = le_ExerciseAngina.fit_transform(df3['ExerciseAngina'])
df3['ST_Slope'] = le_ST_Slope.fit_transform(df3['ST_Slope'])
```

```python
df3.head()
```

```python
df4 = df3.copy()
df4.ExerciseAngina.replace(
    {
        'N': 0,
        'Y': 1
    },
    inplace=True)

df4.ST_Slope.replace(
    {
        'Down': 1,
        'Flat': 2,
        'Up': 3
    },
    inplace=True
)

df4.RestingECG.replace(
    {
        'Normal': 1,
        'ST': 2,
        'LVH': 3
    },
    inplace=True)

df4.head()
```

```python
df5 = pd.get_dummies(df4, drop_first=True)
df5.head()
```

```python
X = df5.drop("HeartDisease",axis="columns")
y = df5.HeartDisease
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
Scaler = StandardScaler()
X_scaled = Scaler.fit_transform(X)
X_scaled
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=30)
```

```python
from sklearn.ensemble import RandomForestClassifier

model_rf = RandomForestClassifier()
model_rf.fit(X_train, y_train)
model_rf.score(X_test, y_test)
```

### Use PCA to reduce dimensions

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=8)
```

```python
X_pca = pca.fit_transform(X)
```

```python
X_train_pca, X_test_pca, y_train, y_test = train_test_split(X_pca, y, test_size=0.2, random_state=30)
```

```python
from sklearn.ensemble import RandomForestClassifier

model_rf = RandomForestClassifier()
model_rf.fit(X_train_pca, y_train)
model_rf.score(X_test_pca, y_test)
```
