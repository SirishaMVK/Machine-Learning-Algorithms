### Dimensinality reduction technique  
Reduces dimensionlity curse


```python
import pandas as pd
from sklearn import datasets
```

```python
digits = datasets.load_digits()
digits
```

```python
digits.keys()
```

```python
digits.data[0].reshape(8,8)
```

```python
df = pd.DataFrame(digits.data, columns=digits.feature_names)
df
```

```python
df["target"] = digits.target
```

```python
df.head()
```

```python
df.describe()
```

```python
X = df
y = df.target
```

### Scale the fetures before building the model

```python
from sklearn.preprocessing import StandardScaler
```

```python
Scaler = StandardScaler()
X_Scaled = Scaler.fit_transform(X)
X_Scaled
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_Scaled, y, test_size=0.2, random_state=30)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
Log_model = LogisticRegression()
Log_model.fit(X_train, y_train)
Log_model.score(X_test,y_test)
```

### PCA

```python
from sklearn.decomposition import PCA
```

```python
#retain 95% of useful info
pcs = PCA(0.95)
```

```python
X_pca = pcs.fit_transform(X)
X_pca
```

```python
X_pca.shape
```

```python
pcs.explained_variance_ratio_
```

```python
pcs.explained_variance_
```

```python
pcs.n_components_
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_pca, y, test_size=0.2, random_state=30)
```

```python
Log_model1 = LogisticRegression(max_iter=100000)
Log_model1.fit(X_train, y_train)
Log_model1.score(X_test,y_test)
```

```python
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)
X_pca.shape
```

```python
X_pca
```

```python
pca.explained_variance_ratio_
```

```python
X_train_pca, X_test_pca, y_train, y_test = train_test_split(X_pca, y, test_size=0.2, random_state=30)

model = LogisticRegression(max_iter=1000)
model.fit(X_train_pca, y_train)
model.score(X_test_pca, y_test)
```
